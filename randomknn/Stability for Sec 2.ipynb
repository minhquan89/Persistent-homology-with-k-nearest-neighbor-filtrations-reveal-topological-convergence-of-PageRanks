#  We create an example for two point clouds (one is created by small perturbation) and compare their peristent diagrams by stability theorem


```python
%pylab inline
#import networkx as nx
import gudhi as gd
from sklearn.metrics import pairwise_distances as dists

import sys
sys.path.append('../CODE')

from NNO_util import *
```

```python
N = 21
thetas = np.linspace(0,2*pi,N)[:-1] + np.random.randn(N-1)/10
X1 = np.cos(thetas)
Y1 = np.sin(thetas)

gamma = .1
X2 = X1 + gamma*np.random.randn(len(X1))
Y2 = Y1 + gamma*np.random.randn(len(Y1))

#X1 = [1,2,3,4,5]
#X2=[0.8,2.2,2.9,4,5.1]
#$Y1 = [4,8,12,16,17]
#Y2 = [4.6,8.8,12.5,16.9,17.5]


#Y2 = [s + gamma*np.random.randn() for s in Y1]
#X2 = [s + gamma*np.random.randn() for s in X1]
#plt.scatter(X1,Y1,color='red')
#plt.scatter(X2,Y2,color='blue')
#plt.show()
```

```python
points_1=list(zip(X1,Y1))
points_1=np.array(points_1)
points_2=list(zip(X2,Y2))
points_2=np.array(points_2)
```

```python
d_1=dists(points_1,points_1)
d_2=dists(points_2,points_2)
diag1=dist2PD(d_1,2,max_edge_length=1.9)[1]
diag2=dist2PD(d_2,2,max_edge_length=1.9)[1]
#diag1,diag2

#d_1-d_2
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))

ax[0].scatter(X1,Y1,color='k')
gd.plot_persistence_diagram(persistence=diag1,axes=ax[1],alpha = .5);
plt.tight_layout()
plt.savefig('perdia.pdf')
```

```python
fig1,ax = plt.subplots(1,3,figsize=(12,4))

ax[0].scatter(X1,Y1,color='k')
ax[0].scatter(X2,Y2,color='y')

gd.plot_persistence_diagram(persistence=diag1,axes=ax[1],alpha = .5);
gd.plot_persistence_diagram(persistence=diag2,axes=ax[2],alpha = .5);

#       if the above doesn't work, update Gudhi 
#     conda update -c conda-forge gudhi 

plt.tight_layout()
plt.savefig('figs/stability_1.pdf')
```
